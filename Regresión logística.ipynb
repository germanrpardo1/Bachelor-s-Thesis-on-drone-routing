```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def logit(v, beta0, beta1):
    y = 1/(1+np.exp(beta0+beta1*v))
    return y
```

```python
beta0 = 6.576
beta1 = -0.092
vrange = range(0,140)
dict_fun = {}

for i in vrange:
    dict_fun[i] = logit(i, beta0, beta1)
    
x = []
z = []

for i in dict_fun.keys():
    x.append(i)
for i in dict_fun.values():
    z.append(i)
    
plt.xlabel('Velocidad de impacto [km/h]')
plt.ylabel('Probabilidad de fatalidad [%]')
plt.title('Regresin logstica de carro contra peatn')
plt.plot(x,z)
```
